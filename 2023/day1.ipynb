```python
def get_list(x):
    l = []
    for i in x:
        if i.isdigit():
            l.append(int(i))
    return l
```

```python
def reformat(x):
    numbers = {
        "one":"o1e",
        "two":"t2o",
        "three":"t3e",
        "four":"f4r",
        "five":"f5e",
        "six":"s6x",
        "seven":"s7n",
        "eight":"e8t",
        "nine":"n9e"
    }
    for key in numbers.keys():
        if x.__contains__(key):
            x=x.replace(key,str(numbers[key]))
    return x
```
